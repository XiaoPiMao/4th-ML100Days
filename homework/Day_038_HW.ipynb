```python
#用 sklearn.datasets.load_wine()
from sklearn import datasets
data = datasets.load_wine()
data
```

```python
# 看一下說明
print(data.DESCR)
```

```python
#做出DataFrame
import pandas as pd
wind_df = pd.DataFrame(data.data, columns=
    ["Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids", 
     "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", ", OD280/OD315 of diluted wines",
     "Proline"])
wind_df
```

```python
#加上traget欄，觀察資料
wind_df["target"] = data.target
wind_df
```

```python
#切分訓練集/測試集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=4)
```

```python
#選擇 Model OneVsRestClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

#訓練模型
ovr_clf = OneVsRestClassifier(SVC()).fit(x_train, y_train)

#預測測試集
y_pred = ovr_clf.predict(x_test)
```

```python
#用accuracy評分
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```
